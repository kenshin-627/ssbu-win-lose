```python
from pynput import mouse
import pyautogui
import datetime
import cv2
import numpy as np
import os
from PIL import Image
# import ctypes

# PROCESS_PER_MONITOR_DPI_AWARE = 2

# ctypes.windll.shcore.SetProcessDpiAwareness(PROCESS_PER_MONITOR_DPI_AWARE)

left_up_x, left_up_y, right_down_x, right_down_y = [0,0,0,0]
num_press = 0

def main():
    with mouse.Listener(on_click=on_click) as listener:
        listener.join()
    # 範囲の指定が終わったらスクリーンショットを撮る
    os.makedirs("screenshot", exist_ok=True)
    # 画像を保存するパスは "screenshot/年月日時分秒.png"
    ss_time = datetime.datetime.now().strftime("%y%m%d%H%M%S")
    path = "screenshot/" + ss_time + ".png"
    
#     im = pyautogui.screenshot("tmp.png", region=(left_up_x*2, left_up_y*2, (right_down_x-left_up_x)*2, (right_down_y-left_up_y)*2))
#     print(type(im))

# -----------------------------------------------------------------
    im_before = cv2.imread("screenshot/before.png")
    im_after = cv2.imread("screenshot/after.png")
    gray_before = cv2.cvtColor(np.array(im_before, dtype=np.uint8), cv2.COLOR_BGR2GRAY)
    gray_after = cv2.cvtColor(np.array(im_after, dtype=np.uint8), cv2.COLOR_BGR2GRAY)
# -----------------------------------------------------------------
#     cv2.imwrite("screenshot/before_gray.png", gray_before)
#     cv2.imwrite("screenshot/after_gray.png", gray_after)
# -----------------------------------------------------------------
    resize_before = cv2.resize(gray_before, dsize=(1024, 576))
    resize_after = cv2.resize(gray_after, dsize=(1024, 576))
# -----------------------------------------------------------------
    
#     im = Image.open("tmp.png")
#     im_gray = im.convert("L")
#     im_gray = np.array(im_gray, dtype=np.uint8)
#     cv2.imwrite("tmp_gray.png", im_gray)
#     im_resize = cv2.resize(im_gray, dsize=(1024, 576))

    lim_down = 130
    lim_up = 400
    
#     im_edges = cv2.Canny(im_resize, lim_down, lim_up, L2gradient=True)
#     cv2.imwrite(f"tmp_edges_{lim_down}-{lim_up}.png", im_edges)
    
# -----------------------------------------------------------------
    edges_before = cv2.Canny(resize_before, lim_down, lim_up, L2gradient=True)
    edges_after = cv2.Canny(resize_after, lim_down, lim_up, L2gradient=True)
    cv2.imwrite(f"screenshot/before_{lim_down}-{lim_up}_edges_gray.png", edges_before)
    cv2.imwrite(f"screenshot/after_{lim_down}-{lim_up}_edges_gray.png", edges_after)
# -----------------------------------------------------------------

    # Hough変換で直線を検出
#     lines_before = cv2.HoughLines(im_edges, rho=1, theta=np.pi/180, threshold=200)
    lines_before = cv2.HoughLines(edges_before, rho=1, theta=np.pi/180, threshold=200)
    # 検出された直線を確認
    print(lines_before)
#     straight_before = cv2.resize(cv2.imread("tmp.png"), dsize=(1024, 576))
    straight_before = cv2.imread("screenshot/before.png")
    if not lines_before is None:
        for line in lines_before:
            rho, theta = line[0]
            draw_line(straight_before, theta, rho)
#         cv2.imwrite("tmp_straight.png", straight_before)
        cv2.imwrite(f"screenshot/before_{lim_down}-{lim_up}_straight.png", straight_before)

def draw_line(img, theta, rho):
    h, w = img.shape[:2]
    # 直線が垂直のとき
    if np.isclose(np.sin(theta), 0):
        x1, y1 = rho, 0
        x2, y2 = rho, h
    # 直線が垂直じゃないとき
    else:
        # 直線の式を式変形すればcalc_yの式がもとまる．
        calc_y = lambda x: rho / np.sin(theta) - x * np.cos(theta) / np.sin(theta)
        x1, y1 = 0, calc_y(0)
        x2, y2 = w, calc_y(w)
    x1, y1, x2, y2 = map(int, [x1, y1, x2, y2])
    # 直線を描画 (imgにアルファチャンネルがあると黒い線分しか引かれなくなる)
    cv2.line(img, (x1, y1), (x2, y2), (0, 255, 0), 2)
    
def on_click(x, y, button, pressed):
    global left_up_x, left_up_y, right_down_x, right_down_y, num_press
    if pressed:
        # 1回目のクリックはウィンドウの切り替え用で何もしない
        if num_press == 0:
            num_press = 1
        # 2回目のクリックはスクリーンショットを撮る範囲の左上を指定
        elif num_press == 1:
            num_press = 2
            left_up_x, left_up_y = x, y
            return (x, y)
        # 3回目のクリックはスクリーンショットを撮る範囲の右下を指定して終了
        else:
            right_down_x, right_down_y = x, y
            return False

if __name__ == "__main__":
#     main()
```
